### CALUCLATING ARITHMETIC MEAN,MEDIAN,MODE USING NUMPY


```python
#importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statistics as stats
import warnings
warnings.filterwarnings("ignore")
```

```python
#creating list and caluclating the mean,median,mode 

lst=[0,8,7,6,5,4,65,63,3,4,78,64,0,33,12,13,12,45,78,90]
# CALUCLATING MEAN
mean=np.mean(lst)
print(mean)

# CALUCLATING MEDIAN
median=np.median(lst)

print(median)

# CALUCLATING MODE
mode=stats.mode(lst)
print(mode)


```

```python
#plotting the graph using matplotlib
x=[5,6,7,9]
y=[76,98,90,76]
plt.plot(x,y)
plt.show()
```

# CALUCLATING ARITHMETIC MEAN,MEDIAN,MODE WITHOUT USING NUMPY

```python
#creating list
lst=[0,8,7,6,5,4,65,63,3,4,78,64,0,33,12,13,12,45,78,90]

#CALUCLATING ARITHMETIC MEAN
mean=sum(lst)/len(lst)
print(mean)

#CALUCLATING MEDIAN
median=len(lst)/2
print(median)
```

```python
# CALUCLATING MODE
lst=[0,8,7,6,5,4,65,63,3,4,78,64,0,33,12,13,12,45,78,90]
from collections import Counter
data = Counter(lst)
# Returns all unique items and their counts
data.most_common()
# Returns the highest occuring term
data.most_common(1) 
```
