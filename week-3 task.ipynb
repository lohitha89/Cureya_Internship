### Encoding the Data

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
df['fruit_subtype']=le.fit_transform(df['fruit_subtype'])
```

```python
df
```

### Data Standardization

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
```

```python
scaled=ss.fit_transform(df.drop(columns=['fruit_name','fruit_label']))
scaled
```

```python
df_ss=pd.DataFrame(scaled)
df_ss
```

```python
df_12=pd.concat([df_ss,df['fruit_label']],axis=1)
```

```python
df_12
```

### Splitting the Data

```python
x=df_12.drop(columns=['fruit_label'])
y=df_12['fruit_label']
```

```python
 from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.2,random_state=1)
```

### Creating a Model

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier()
```

```python
knn.fit(x_train,y_train)
```

```python
y_pred=knn.predict(x_test)
y_train_pred=knn.predict(x_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_train_pred,y_train)
```

```python
accuracy_score(y_pred,y_test)
```
